## Feature Engineering: independent variable

```python
dataset.dtypes
```

```python
dataset["first_open"] = [parser.parse(row_date) for row_date in dataset["first_open"]]
#only parse string, not NaN
dataset["enrolled_date"] = [parser.parse(row_date) if isinstance(row_date, str) else row_date for row_date in dataset["enrolled_date"]]
```

```python
dataset.dtypes
```

```python
dataset.head(5)
```

#### calculate first open and enroll date difference in hour

```python
dataset["difference"] = (dataset.enrolled_date-dataset.first_open).astype('timedelta64[h]')
```

#### investigate the difference in histogram

```python
dataset.head(10)
```

```python
response_hist = plt.hist(dataset["difference"].dropna(), color='#3F5D7D')
plt.title('Distribution of Time Difference between Installation and Enrollment')
plt.show()
```

#### investigate the difference in histogram in first 100 hours

```python
response_hist = plt.hist(dataset["difference"].dropna(), color='#3F5D7D', range = [0, 100])
plt.title('Distribution of Time Difference (first 100 h)')
plt.show()
```

### take enrollment cut off date as first 48 hours (2 days)

```python
dataset.loc[dataset.difference > 48, 'enrolled'] = 0
dataset = dataset.drop(columns=['enrolled_date', 'difference', 'first_open'])
```

## Feature Engineering: dependent variable: screen_list

```python
# Load Top Screens
top_screens = pd.read_csv('top_screens.csv').top_screens.values
top_screens
```

```python
dataset["screen_list"].iloc[0]
```

### create a column for each most popular screens

```python
# Mapping Screens to Fields
dataset["screen_list"] = dataset.screen_list.astype(str) + ','
for sc in top_screens:
    dataset[sc] = dataset.screen_list.str.contains(sc).astype(int)
    dataset['screen_list'] = dataset.screen_list.str.replace(sc+",", "")
```

```python
dataset.head()
```

### create other column as number of unpopular columns

```python
dataset['Other'] = dataset.screen_list.str.count(",")
dataset = dataset.drop(columns=['screen_list'])
```

```python
dataset.head()
```

### combine saving columns into one

```python
savings_screens = ["Saving1",
                    "Saving2",
                    "Saving2Amount",
                    "Saving4",
                    "Saving5",
                    "Saving6",
                    "Saving7",
                    "Saving8",
                    "Saving9",
                    "Saving10"]
dataset["SavingCount"] = dataset[savings_screens].sum(axis=1)
dataset = dataset.drop(columns=savings_screens)
```

### combine credit columns into one

```python
cm_screens = ["Credit1",
               "Credit2",
               "Credit3",
               "Credit3Container",
               "Credit3Dashboard"]
dataset["CMCount"] = dataset[cm_screens].sum(axis=1)
dataset = dataset.drop(columns=cm_screens)
```

### combine cc columns into one

```python
cc_screens = ["CC1",
                "CC1Category",
                "CC3"]
dataset["CCCount"] = dataset[cc_screens].sum(axis=1)
dataset = dataset.drop(columns=cc_screens)
```

### combine loan columns into one

```python
loan_screens = ["Loan",
               "Loan2",
               "Loan3",
               "Loan4"]
dataset["LoansCount"] = dataset[loan_screens].sum(axis=1)
dataset = dataset.drop(columns=loan_screens)
```
